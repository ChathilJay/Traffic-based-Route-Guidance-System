```python
from tensorflow.keras.models import load_model
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
from Volume_predict import predict_volume
```

```python
df = pd.read_csv("../Documents/traffic_dataset_cleaned.csv")
df["DateTime"] = pd.to_datetime(df["DateTime"])
```

```python
scaler = MinMaxScaler()
scaler.fit(df["Volume"].values.reshape(-1, 1))
```

```python
model = load_model("LSTM_model.h5",compile=False)
```

```python
from tensorflow.keras.losses import MeanSquaredError


model.compile(optimizer='adam', loss=MeanSquaredError())
```

```python
df[df["SCATS Number"] == 2000].sort_values("DateTime").head(30)
```

```python
volume = predict_volume(df, site_id=2000, model=model, scaler=scaler, window_size=24, start_time_str="2006-10-08 23:00:00")
print("Predicted volume:", volume)
```
